### Зачем нужен годограф Найквиста?

1. **Анализ устойчивости**:
   - **Открытые контуры**: Годограф Найквиста используется для анализа устойчивости системы с открытым контуром. Он помогает понять, как ведет себя система при различных частотах.
   - **Устойчивость замкнутого контура**: При помощи критерия Найквиста можно определить, будет ли система с замкнутым контуром устойчива, основываясь на годографе передаточной функции открытого контура.

2. **Критерий Найквиста**:
   - Позволяет определить количество нестабильных корней характеристического уравнения замкнутой системы.
   - Критерий гласит, что система замкнутого контура устойчива, если годограф функции $ L(s) $ не охватывает точку $ -1 $ или охватывает её необходимое количество раз в зависимости от количества полюсов открытого контура в правой полуплоскости.

3. **Частотный анализ**:
   - Годограф показывает, как изменяется амплитуда и фаза выходного сигнала при различных частотах входного сигнала. Это позволяет выявить резонансные частоты и частоты, при которых система испытывает серьезные фазовые сдвиги.

### Примеры использования:

- **Авиационная и автомобильная техника**: анализ устойчивости автопилота, систем поддержки водителя.
- **Энергетика**: анализ устойчивости регуляторов частоты в генераторах.
- **Производство**: анализ устойчивости ПИД-регуляторов в технологических процессах.

### Пример RC-фильтра:
   - Сопротивление $ R $
   - Емкость $ C $

**Составим уравнение с использованием закона Кирхгофа для напряжения**:
$$ V_{in}(s) = I(s)R + \frac{I(s)}{sC} $$
$$ I(s) = \frac{V_{out}(s)}{sC} $$

$$ V_{in}(s) = sRC \cdot V_{out}(s) + V_{out}(s) $$
$$ H(s) = \frac{V_{out}(s)}{V_{in}(s)} = \frac{1}{sRC + 1} $$

```julia
import Pkg
Pkg.add("ControlSystems", io=devnull)
Pkg.add("Plots", io=devnull)
Pkg.add("DSP", io=devnull)
```

```julia
using ControlSystems
using DSP
using Plots

# Параметры RC-фильтра
R = 1000  # Ом
C = 0.000001  # Фарад

# Создаем передаточную функцию H(s) = 1 / (RC*s + 1)
num = [1.0]  # числитель
den = [R*C, 1.0]  # знаменатель

# Передаточная функция
sys = tf(num, den)

# Создание логарифмически распределенных частот
ω = exp10.(range(-6, stop=6, length=1000))  # частоты от 10^-6 до 10^6 рад/с

# Получаем данные для построения графика Найквиста
response = ControlSystems.freqresp(sys, ω)

# Функция для построения годографа Найквиста
function plot_nyquist(response)
    p = plot(
        real(response[1,1,:]), 
        imag(response[1,1,:]), 
        label="Nyquist Plot", 
        xlabel="Real", 
        ylabel="Imaginary", 
        legend=:topright, 
        title="Nyquist Plot", 
        aspect_ratio=:equal
    )
    scatter!([-1], [0], label="-1 point", markershape=:star, markersize=6, markercolor=:red)
    display(p)
end

# Построение и отображение графика
plot_nyquist(response)

println("График Найквиста построен. Проверьте, охватывает ли он точку -1.")
```

Как мы видим, график не охватывает точку -1 $ \Rightarrow $ система устойчива
